```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Define the ticker symbol for TSMC (Taiwan Semiconductor Manufacturing Company)
ticker_symbol = '2330.TW'

# Fetch historical data for TSMC for the past year
start_date = pd.to_datetime('today') - pd.DateOffset(years=1)
df = yf.download(ticker_symbol, start=start_date.strftime('%Y-%m-%d'), end=pd.to_datetime('today').strftime('%Y-%m-%d'))

# Display the first few rows of the data
print("Historical data for TSMC (Past Year):")
display(df.head())

# Plot the closing price
plt.figure(figsize=(12, 6))
plt.plot(df['Close'])
plt.title('TSMC (2330.TW) Stock Price Trend (Past Year)')
plt.xlabel('Date')
plt.ylabel('Closing Price (TWD)')
plt.grid(True)
plt.show()
```
